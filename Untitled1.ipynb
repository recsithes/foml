```python
import re
from collections import Counter

def process_data(file_name):
    with open(file_name, 'r') as f:
        lines = f.readlines()
        words = []
        for line in lines:
            line = line.strip().lower()
            word_list = re.findall(r'\w+', line)
            words.extend(word_list)
    return words

words = process_data('/content/dataset.txt')
vocab = set(words)
word_counts = Counter(words)

print(f'There are {len(words)} total words in the corpus.')
print(f'There are {len(vocab)} unique words in the vocabulary.')
```

```python
def get_probs(word_counts):
    probs = {}
    m = sum(word_counts.values())
    for word in word_counts:
        probs[word] = word_counts[word] / m
    return probs

def edit_one_letter(word, allow_swaps=True):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    splits = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    deletes = [L + R[1:] for L, R in splits if R]
    swaps = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1] if allow_swaps else []
    replaces = [L + c + R[1:] for L, R in splits if R for c in letters]
    inserts = [L + c + R for L, R in splits for c in letters]
    return set(deletes + swaps + replaces + inserts)

def edit_two_letters(word, allow_swaps=True):
    edit_one_set = edit_one_letter(word, allow_swaps)
    edit_two_set = set()
    for w in edit_one_set:
        edit_two_set.update(edit_one_letter(w, allow_swaps))
    return edit_two_set

def get_corrections(word, probs, vocab, n=2):
    # 1. If word is in vocab, suggest the word itself
    # 2. Otherwise, suggest words 1 edit away
    # 3. Otherwise, suggest words 2 edits away
    # 4. Otherwise, return the word itself
    suggestions = (word in vocab and {word}) or \
                  edit_one_letter(word).intersection(vocab) or \
                  edit_two_letters(word).intersection(vocab) or \
                  {word}

    best_guesses = sorted([[s, probs.get(s, 0)] for s in suggestions], key=lambda x: x[1], reverse=True)
    return best_guesses[:n]

# Example usage with 'goagle' (targeting 'google')
probs = get_probs(word_counts)
my_word = 'goagle'
corrections = get_corrections(my_word, probs, vocab)

print(f"Suggestions for '{my_word}':")
for i, word_prob in enumerate(corrections):
    print(f'word {i}: {word_prob[0]}, probability {word_prob[1]:.6f}')
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Create text input and output widgets
input_text = widgets.Text(
    value='',
    placeholder='Type a word...',
    description='Input:',
    disabled=False
)

output = widgets.Output()

def on_value_change(change):
    with output:
        clear_output(wait=True)
        word = change['new'].strip().lower()
        if word:
            # Get up to 5 suggestions for better variety
            suggestions = get_corrections(word, probs, vocab, n=5)
            print(f"Suggestions for '{word}':")
            for i, (suggestion, prob) in enumerate(suggestions):
                print(f"{i+1}. {suggestion} (score: {prob:.6f})")
        else:
            print("Start typing to see suggestions...")

# Observe changes in the text box
input_text.observe(on_value_change, names='value')

# Display the widgets
display(input_text, output)
```
